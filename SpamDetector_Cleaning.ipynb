```python
import pandas as pd
```

```python
df = pd.read_csv('spam.csv', encoding='latin1')

df



```

```python
df = df.isnull().sum()

```

```python
df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'])
```

```python
df = df[df['v1'] == 'spam']

df_cleaned = df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'])

df_cleaned.reset_index

df_cleaned = df_cleaned.rename(columns={'v1': 'Category', 'v2': 'Content'})


```

```python
print(df_cleaned.columns.tolist())

df_final = df_cleaned.reset_index(drop=True)

df_final

```

```python
df_final.to_csv('spam_data_cleaned.csv', index=False)
```
