```python
import pandas as pd
```

```python
spam = pd.read_csv('spam.csv')

spam_1 = spam.drop(columns=['Unnamed: 2'	,'Unnamed: 3',	'Unnamed: 4'])



```

```python
spam_1.drop_duplicates

spam_1.dropna
```

```python
spam_1['v2'] = spam_1['v2'].str.lower()

import re
spam_1['v2'] = spam_1['v2'].apply(lambda x: re.sub('[^A-Za-z0-9]', ' ', x))
```

```python
spam_2 = spam_1[spam_1['v1'] == 'spam']


```

```python
spam_2.reset_index(drop=True)
```

```python
spam_2 = spam_2.rename(columns={'v1': 'Category', 'v2': 'Content'})
```

```python
spam_2.to_csv('spam_data.csv')
```
