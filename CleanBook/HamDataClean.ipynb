```python
import pandas as pd 

df_ham = pd.read_csv('spam.csv')

df_ham
```

```python
df_ham.drop_duplicates()


df_ham0 = df_ham.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'])

df_ham0
```

```python
import re
df_ham0['v2'] = df_ham0['v2'].apply(lambda x: re.sub('[^A-Za-z0-9]', ' ', x))


```

```python
df_ham0['v2'] = df_ham0['v2'].str.lower()
```

```python
df_ham0
```

```python
df_ham04 = df_ham0.rename(columns={'v1': 'Category', 'v2': 'Content'})
```

```python
df = df_ham04[df_ham04['Category'] == 'ham']
```

```python
df.to_csv('spam_data_cleaned.csv', index=False)
```
